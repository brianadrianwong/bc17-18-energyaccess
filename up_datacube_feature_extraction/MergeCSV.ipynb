```python
import glob, os
import pandas as pd
```

```python
# FUNCTIONS
def joinDFbyID(list_of_dfs):
    return pd.concat(list_of_dfs)\
      .groupby(['system:index', 'system_ind'], as_index=False, sort=False).first()

def removeColumns(list_of_cols, df):
    return df.drop(list_of_cols, axis=1)

def extractTag(filename):
    return filename.split('_')[-2]

def addSuffixToCol(df, suffix):
    return df.add_suffix(suffix)\
        .rename(index=str, 
                columns={"system:index"+suffix:"system:index", 
                         "system_ind"+suffix:"system_ind",
                         ".geo"+suffix:".geo"})
```

```python
myCSVdir = 'up_5km'
raw_df = []
cols_to_remove = ['.geo']
for file in os.listdir(myCSVdir):
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(myCSVdir, file))
        if (df.shape[0] != 11402):
            print(file)
        tag = extractTag(file)
        raw_df.append(removeColumns(cols_to_remove, addSuffixToCol(df, '_'+tag)))
raw_df[0].head()
```

```python
cols_to_remove_2 = ['01_03_GREEN_sum', '01_03_NDVI_sum', 
                  '04_06_GREEN_sum', '04_06_NDVI_sum', 
                  '07_09_GREEN_sum', '07_09_NDVI_sum',
                  '10_12_GREEN_sum', '10_12_NDVI_sum']
merged_df = joinDFbyID(raw_df)
print(merged_df.shape[0])
final_df = removeColumns(cols_to_remove_2, merged_df)
final_df.head()
# list(merged_df)
```

```python
final_df = final_df.reindex_axis(sorted(final_df.columns), axis=1)
final_df.head()
list(final_df)
# print(final_df.shape[0])
```
